```python
class cross_val:
    # number of trained models is K1(K2L + 1)
    def trained_models(self, k1, k2, S, show_res=True):
        """
        Calculates number of models trained in a cross validation with k1 inner folds and k2 outer folds
        k1 = number of outer folds
        k2 = number of inner folds
        S = number of models
        show_res = whether to show th final amount of trained models
        """

        runs = 0
        pr_model = 0
        for _ in range(k1):
            for _ in range(k2):
                pr_model += 1
                for _ in range(S):
                    runs += 1
            runs+=1

        if show_res:
            print(
                "A total of {} models were trained, as {} models were trained {} times each. Plus an extra {} in the outer loop, once params are chosen".format(
                    runs, S, pr_model, k1
                )
            )

        return runs
```
