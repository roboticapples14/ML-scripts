Get Euclidian Norm

```python
from numpy import array
from numpy.linalg import norm
arr = array([0.45, -.4, .58, .55])
arr2 = array([-0.6, -.8, -.01, -.08])
print(arr)
# euclidian norm
euclidian_norm = norm(arr, ord=2)
print(euclidian_norm)
# Frobenius_norm (for matrix)
Frobenius_norm = norm([arr], 'fro')
print(Frobenius_norm)
```
