```python
import math

# try out different values of x

def test_regression_weights(x, w0, w1):
    return 1/(1+math.pow(math.e, -(w0 + x * w1)))

# y should be 0 when x = 18
test_0 = test_regression_weights(18, 418.94, -26.12)
# y should be 1 when x = 14
test_1 = test_regression_weights(14, 418.94, -26.12)
```
